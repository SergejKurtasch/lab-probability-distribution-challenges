# Challenge 1 - Plot Probability Mass Function (PMF)

#### Background knowledge

[PMF](https://en.wikipedia.org/wiki/Probability_mass_function) shows the probability distribution of a **discrete random variable**. A [discrete random variable](https://en.wikipedia.org/wiki/Random_variable#Discrete_random_variable) has random numeric values that are not continuous. For example, the number of people in a household can only be integers but not floats. Therefore the number of people in a household is a discrete variable.

#### Question: We assume that the probability of clicking an Ad in a Youtube video is 0.15. We have a sample of 5 people who wathched the video and we want to plot the PMF for a binomial distribution.

*Hint: use binom from `scipy.stats.binom`. Your output should look like below:*

![binom 10](ch-2.png)

```python
# Your code here
p = 0.15
n = 5
x = np.arange(0, n + 1)
# Calculate the binomial probability mass function
pmf = stats.binom.pmf(x, n, p)
# Plot the PMF
plt.figure(figsize=(8, 5))

plt.plot(x, pmf,  ms=8, label='PMF')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF (n=5, p=0.15)')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

#### Explain what you observe from the plot above

```python
# Your comment here
for i in range (0, n + 1):
    print(f'P(X={i}) = {pmf[i]:.4f}')

print(sum(pmf))

# The plot shows the probability of getting 0 to 5 successes in 5 trials with a success probability of 0.15. The highest probability is for 0 successes, which is expected since the success probability is low. The probabilities decrease as the number of successes increases, reflecting the nature of the binomial distribution with a low success probability.
# The sum of the probabilities is 1, confirming that the PMF is correctly normalized.
```

#### Reaserch the Poisson distribution. Write about your own understanding of the Poisson distribution.

```python
# Your comment here
# The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, given that these events occur with a known constant mean rate and independently of the time since the last event.
```

#### A website has an average of 300 visits per day. What is the probability of getting 320 visitors in a day?

*Hint: use `scipy.stats.poisson`*

```python
# Your code here
lam = 300
n = 320
x = np.arange(200, 400, 5)
# Calculate the Poisson probability mass function
pmf = stats.poisson.pmf(x, lam)
# Plot the PMF
plt.figure(figsize=(16, 5))
plt.plot(x, pmf,  ms=8, label='PMF')
plt.axvline(x=320, color='r', linestyle='--', label='λ')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution PMF (λ={lam})')
plt.xticks(x, rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

p_320 = stats.poisson.pmf(320, lam)
print(f'P(X=320) = {p_320:.5f}')
```

#### What is the probability of getting 60 visits?

```python
# Your code here
pmf_60 = stats.poisson.pmf(60, lam)
print(f'P(X=60) = {pmf_60:.9f}')
```
